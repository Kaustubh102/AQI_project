## Extracting the pollutants

```python
pm2=[]
pm10=[]
co=[]
so2=[]
no2=[]
o3=[]
aqi=[]


for x, y in zip(telangana['lat'],telangana['long']):
    # Make the request
    response = requests.get(
        f'https://www.aqi.in/air-quality-map?type=aqi&lat={x}&lon={y}',
        verify=False
    )
    response.text

    l1=re.findall(r'(\d+)\sµg/m³',response.text)
    pm2.append(l1[0])
    pm10.append(l1[1])

    l2=re.findall(r'(\d+)\sppb',response.text)
    co.append(l2[0])
    so2.append(l2[1])
    no2.append(l2[2])
    o3.append(l2[3])
    
    l3=re.findall('\B>(\d+)<\B',response.text)
    aqi.append(l3[0])
```

```python
telangana['AQI']=aqi
telangana['pm2.5']=pm2
telangana['pm10']=pm10
telangana['co']=co
telangana['so2']=so2
telangana['no2']=no2
telangana['o3']=o3
```

```python
telangana
```

## Extracting the temp wind speed and humidity

```python
wind_speed=[]
humdity=[]
pressure=[]
temperature=[]



for x,y in zip(telangana['lat'],telangana['long']):
    url = "https://api.weatherstack.com/current?access_key=5d81b6f7989baf27c643117e6a60abac"

    querystring = {"query":f"{x},{y}"}

    response = requests.get(url, params=querystring)
    wind_speed.append(re.findall('wind_speed\':\s(\d+\.?\d*)',str(response.json()))[0])
    humdity.append(re.findall('humidity\':\s(\d+\.?\d*)',str(response.json()))[0])
    pressure.append(re.findall('pressure\':\s(\d+\.?\d*)',str(response.json()))[0])
    temperature.append(re.findall('temperature\':\s(\d+\.?\d*)',str(response.json()))[0])
```

```python
telangana['temperature']=temperature
telangana['wind_speed']=wind_speed
telangana['humdity']=humdity
telangana['pressure']=pressure
```
